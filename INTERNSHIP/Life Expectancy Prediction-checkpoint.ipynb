#### Filtering Outliers

```python
lifetime.head()
```

```python
def outliers(x):
    q1 = np.percentile(x,25)
    q3 = np.percentile(x,75)
    iqr = q3-q1
    floor = q1 - 1.5*iqr
    ceiling = q3 + 1.5*iqr
    leng =len(x[(x<floor)|(x>ceiling)])
    return leng
```

```python
for col in lifetime.columns[3:]:
    print('{:35} : {:5}'.format(col,outliers(lifetime[col])))
```

```python
from scipy.stats.mstats import winsorize
```

```python
def outliers_left_right_percentage(x):
    leng = x.shape[0]
    q1 = np.percentile(x,25)
    q3 = np.percentile(x,75)
    iqr = q3-q1
    floor = q1 - 1.5*iqr
    ceiling = q3 + 1.5*iqr
    left = len(x[x<floor])/leng
    right = len(x[x>ceiling])/leng
    return (left,right)
```

```python
for columns in lifetime.columns[3:]:
    print('{:35} : {:.5f},  {:.5f} '.format(columns,*outliers_left_right_percentage(lifetime[columns])))
```

```python
new_data = pd.DataFrame(lifetime[['Country','Year','Status']])
```

```python
for col in lifetime.columns[3:]:
    new_data[col] = winsorize(lifetime[col],outliers_left_right_percentage(lifetime[col]))
    print('{:35} : {:3},  {:3}'.format(col,outliers(new_data[col]),outliers(lifetime[col])))
```

#### spliting data into two sets(train,test)

```python
x = new_data.drop(['Life expectancy'],axis = 1)
y = new_data['Life expectancy']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.25,random_state = 45)
```

```python
print("X_train  shape",x_train.shape)
print("Y_train  shape",y_train.shape)
print("X_test   shape",x_test.shape)
print("X_test   shape",y_test.shape)
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder

c_tranform = Pipeline(steps = [('onehot',OneHotEncoder(handle_unknown = 'ignore')),])
```

```python
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler

n_tranform = Pipeline(steps=[('imputer',SimpleImputer(strategy = 'median')),])
```

```python
from sklearn.compose import ColumnTransformer

pre_process = ColumnTransformer(transformers=[
    ('cat',c_tranform,new_data.columns[:3]),
    ('num',n_tranform,new_data.columns[4:])
])
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score
```

```python
lm = LinearRegression()
dt = DecisionTreeRegressor()
rf = RandomForestRegressor()
```

```python
pip_lm = Pipeline([('preprocessor',pre_process),('RFRegressor',lm)])
pip_lm.fit(x_train,y_train)
```

```python
pip_dt = Pipeline([('preprocessor',pre_process),('RFRegressor',dt)])
pip_dt.fit(x_train,y_train)
```

```python
pip_rf = Pipeline([('preprocessor',pre_process),('RFRegresssor',rf)])
pip_rf.fit(x_train,y_train)
```

```python
lm_predict = pip_lm.predict(x_test)
dt_predict = pip_dt.predict(x_test)
rf_predict = pip_rf.predict(x_test)
```
